```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
def fatorial(n):
    if n<=1:
        return 1
    else :
        return fatorial(n-1)*n
```

```python
def T(n):
    if n<=1: 
        return 1
    else:
        return T(n-1)+1
```

```python
def busca(x, l, inicio, fim):
    if(inicio>=fim):
        return -1
    else:
        mid = (inicio+fim)//2
        if x == l[mid]:
            return mid
        elif x<l[mid]:
            return busca(x,l,inicio, mid-1)
        else:
            return busca(x,l, mid+1, fim)
```

```python
def T(n):
    if n==1:
        return 1
    else:
        return T(n//2)+1
```

```python
T(3)
```

```python
Tanterior = T(1)
for i in range(2,100):
    Tatual = T(i)
    plt.plot((i-1,i),(Tanterior,Tatual),'-')
    Tanterior = Tatual
x = range(2,100)
plt.plot(x, np.log2(x))
```

```python
def T(n):
    if n==1:
        return 1
    else:
        return 2*T(n//2)+n
```

```python
Tanterior = T(1)
for i in range(2,100):
    Tatual = T(i)
    plt.plot((i-1,i),(Tanterior,Tatual),'-')
    Tanterior = Tatual
x = range(2,100)
plt.plot(x, np.log2(x)*x)
```

```python
def T(n):
    if n==1:
        return 1
    else:
        return 3*T(n//2)+n*math.log2(n)
```

```python
Tanterior = T(1)
for i in range(2,1000):
    Tatual = T(i)
    plt.plot((i-1,i),(Tanterior,Tatual),'-')
    Tanterior = Tatual
x = range(2,1000)
plt.plot(x, 4*np.power(x, math.log2(3)))
```
